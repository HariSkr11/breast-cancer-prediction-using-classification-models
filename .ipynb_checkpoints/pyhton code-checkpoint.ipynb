## Analysis of each model

### Logistic Regression

logistic regression is a type of regression model in which we are fitting the data into a sigmoid function instead of a linear function as we did in linear regression

image.png

>So what we are doing is that for linear regression the function was $y_{x}=mx+c$

>But when it comes to logistic regression the function will be

>$y_{x}=\frac{1}{1+e^{(mx+c)}}$ 

```python
lg = LogisticRegression() #Using this will change the accuracy score 
lg.fit(X_train,y_train)
print(lg.score(X_test,y_test))
```

## Applying K fold cross validation to improve the accuracy of our models

```python
data.columns
```

```python
kf = KFold(n_splits=10)
input_data=(data.reset_index()).drop(['diagnosis','diagnosis_target','id'],axis='columns')
out_data = data['diagnosis']
allmodels = [rf,lg,dc,sv]
```

```python
def get_best_model_kf(kf,model,input_data,out_data):
   model_list=[]
   score_list=[]
   for train_index,test_index in kf.split(input_data):
       X_train,X_test,y_train,y_test = input_data.iloc[train_index],input_data.iloc[test_index],out_data.iloc[train_index],out_data.iloc[test_index]
       model.fit(X_train,y_train)
       model_list.append(model)
       score_list.append(model.score(X_test,y_test))
   best_score = max((score_list))
   index_number = score_list.index(float(max(score_list)))
   best_model = model_list[index_number]
   return best_model,best_score
```

```python
model_list=['Random forest','Logistic regression','Decision tree classifier','Support vector classifier']
model_content=[get_best_model_kf(kf,i,input_data,out_data)[0] for i in allmodels]
accuracy_content = [get_best_model_kf(kf,i,input_data,out_data)[1]*100 for i in allmodels]
```

```python
for i,d in enumerate(model_list):
    print("The accuracy score of the model",d,'is',accuracy_content[i])
```

```python
sns.set(rc={'figure.figsize':(15,15)})
ax = sns.barplot(x=model_list,y=accuracy_content,palette="Blues_d")
for rect in ax.patches:
    ax.text (rect.get_x() + rect.get_width()  / 2,rect.get_height()+ 0.75,rect.get_height(),horizontalalignment='center', fontsize = 12)
```

```python
def best_kf_model_getter(start_index,stop_index,model):
    kf_names = ['Kf '+str(i) for i in range(start_index,stop_index+1)]
    kf_list=[KFold(n_splits=i) for i in range(start_index,stop_index+1)]
    acc_list = [get_best_model_kf(kf,model_content[2],input_data,out_data)[1]*100 for kf in kf_list]
    a = max(acc_list)
    b = kf_names[acc_list.index(a)]
    return a,b
```
